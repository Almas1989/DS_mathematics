```python
import numpy as np
import matplotlib.pyplot as plt

# Коэффициенты из твоего примера: f(x) = x³ - 6x² + 9x + 15
a, b, c, d = 1, -6, 9, 15
p, q = 0, 5
```

Математическое объяснениеПрежде чем писать код, давай разберем логику:Функция: $f(x) = ax^3 + bx^2 + cx + d$ — это "ландшафт".
1. Первая производная ($f'$): $3ax^2 + 2bx + c$. Она показывает скорость изменения функции. В точках экстремума (пики и впадины) скорость равна 0.
2. Вторая производная ($f''$): $6ax + 2b$. Она показывает ускорение (выпуклость).
   Если $f'' < 0$, график "грустный" (выпуклый вверх) $\cap$ — это максимум.
   Если $f'' > 0$, график "улыбается" (выпуклый вниз) $\cup$ — это минимум.
   Если $f'' = 0$, это точка перехода (перегиб).

```python
def f(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d

def df(x, a, b, c, d):
    return 3*a*x**2 + 2*b*x + c

# Создаем массив точек для графика
x_vals = np.linspace(p, q, 400)
y_vals = f(x_vals, a, b, c, d)
dy_vals = df(x_vals, a, b, c, d)

# Визуализация
plt.figure(figsize=(10, 6))

# Рисуем саму функцию
plt.plot(x_vals, y_vals, label='f(x) = x³ - 6x² + 9x + 15', color='blue', lw=2)

# Рисуем производную (чтобы видеть, где она пересекает 0)
plt.plot(x_vals, dy_vals, label="f'(x) (производная)", color='orange', linestyle='--')

# Отмечаем критические точки из примера (1 и 3)
critical_x = [1, 3]
critical_y = [f(1, a, b, c, d), f(3, a, b, c, d)]

plt.scatter(critical_x, critical_y, color='red', zorder=5)
plt.annotate('Local Max (1, 19)', xy=(1, 19), xytext=(1.2, 21),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Local Min (3, 15)', xy=(3, 15), xytext=(3.2, 13),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Оформление
plt.axhline(0, color='black', lw=1)
plt.title("Анализ функции и её критических точек")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def analyze_poly(a, b, c, d, p, q, title):
    x = np.linspace(p, q, 400)
    
    # f(x), f'(x), f''(x)
    y = a*x**3 + b*x**2 + c*x + d
    dy = 3*a*x**2 + 2*b*x + c
    ddy = 6*a*x + 2*b
    
    plt.figure(figsize=(12, 5))
    
    # Основной график
    plt.subplot(1, 2, 1)
    plt.plot(x, y, label='f(x) - Функция', color='blue', lw=2)
    plt.axhline(0, color='black', alpha=0.3)
    plt.title(f"Функция: {title}")
    plt.legend()
    plt.grid(True, alpha=0.2)
    
    # График производных
    plt.subplot(1, 2, 2)
    plt.plot(x, dy, label="f'(x) - Скорость", color='orange', lw=2)
    plt.plot(x, ddy, label="f''(x) - Ускорение", color='green', linestyle=':')
    plt.axhline(0, color='red', lw=1, label="0-линия")
    plt.title("Анализ производных")
    plt.legend()
    plt.grid(True, alpha=0.2)
    
    plt.tight_layout()
    plt.show()

# Пример 1: Твой пример (Max и Min)
analyze_poly(1, -6, 9, 15, 0, 4, "С двумя экстремумами")

# Пример 2: Точка перегиба (Saddle Point)
# В x=0 и f' и f'' равны нулю!
analyze_poly(1, 0, 0, 0, -2, 2, "Точка перегиба (x^3)")

# Пример 3: Постоянный рост
analyze_poly(1, 0, 3, 0, -2, 2, "Без критических точек")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): return 1*x**2 - 2*x - 3
def df(x): return 2*x - 2

x_val = 4.0  # Начальная точка
iterations = 3
x_range = np.linspace(-2, 5, 400)

plt.figure(figsize=(10, 6))
plt.plot(x_range, f(x_range), label='f(x) = x² - 2x - 3', color='blue')
plt.axhline(0, color='black', lw=1)

for i in range(iterations):
    y_val = f(x_val)
    slope = df(x_val)
    
    # Рисуем точку на кривой
    plt.scatter([x_val], [y_val], color='red')
    
    # Рисуем касательную: y = slope*(x - x_val) + y_val
    x_tangent = np.array([x_val - 1, x_val + 1])
    # Чтобы касательная доходила до оси X:
    next_x = x_val - y_val / slope
    plt.plot([x_val, next_x], [y_val, 0], '--', color='orange', alpha=0.6)
    
    print(f"Итерация {i+1}: x = {x_val:.4f}, f(x) = {y_val:.4f}")
    x_val = next_x

plt.title("Визуализация шагов метода Ньютона")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные примера
g = np.array([2.0, 0.0])
v1 = np.array([1.0, 0.0])
v2 = np.array([0.0, 1.0])
v3 = np.array([-1.0, 0.0])
epsilon = 0.5

directions = [v1, v2, v3]
colors = ['purple', 'green', 'red']

plt.figure(figsize=(8, 8))

# Рисуем вектор градиента (синий)
plt.quiver(0, 0, g[0], g[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Градиент g (направление роста)')

# Рисуем предложенные шаги (учитывая нормирование и epsilon)
for i, v in enumerate(directions, 1):
    norm_v = np.linalg.norm(v)
    unit_v = v / norm_v
    step = epsilon * unit_v
    
    # Считаем изменение
    delta_L = np.dot(g, step)
    
    plt.quiver(0, 0, step[0], step[1], angles='xy', scale_units='xy', scale=1, 
               color=colors[i-1], label=f'Шаг v{i} (ΔL = {delta_L:.1f})')

# Оформление графика
plt.xlim(-1.5, 2.5)
plt.ylim(-1.5, 1.5)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid(True, alpha=0.3)
plt.title("Выбор направления шага по линейному приближению")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Зафиксируем веса и целевой y
w1, w2, y = 0.5, 3.0, 4.0

# Создадим диапазон входных значений x от -5 до 5
x_vals = np.linspace(-5, 5, 500)
dw1_vals = []
dw2_vals = []
loss_vals = []

for x in x_vals:
    # Forward
    a = w1 * x
    h = max(0.0, a)
    y_pred = w2 * h
    loss = 0.5 * (y_pred - y)**2
    
    # Backward
    dL_dy_pred = y_pred - y
    dL_dw2 = dL_dy_pred * h
    dL_dh = dL_dy_pred * w2
    dh_da = 1.0 if a > 0 else 0.0
    dL_da = dL_dh * dh_da
    dL_dw1 = dL_da * x
    
    loss_vals.append(loss)
    dw1_vals.append(dL_dw1)
    dw2_vals.append(dL_dw2)

# Визуализация
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_vals, loss_vals, label='Функция потерь (Loss)', color='black')
plt.title("Зависимость ошибки от входа $x$")
plt.xlabel("Входное значение x")
plt.ylabel("Loss")
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(x_vals, dw1_vals, label=r'$\partial L / \partial w_1$', color='red', lw=2)
plt.plot(x_vals, dw2_vals, label=r'$\partial L / \partial w_2$', color='blue', linestyle='--')
plt.axvline(0, color='gray', linestyle=':', alpha=0.7, label='Граница ReLU (a=0)')
plt.title("Чувствительность весов (Градиенты)")
plt.xlabel("Входное значение x")
plt.ylabel("Значение градиента")
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
