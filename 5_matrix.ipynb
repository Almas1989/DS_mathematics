```python
import numpy as np
from scipy.sparse import coo_matrix

# Определяем ненулевые элементы: их значения и позиции (строки и столбцы)
rows = np.array([0, 1, 2, 0])
cols = np.array([1, 2, 0, 2])
data = np.array([10, 20, 30, 40])

# Создаём матрицу размером 3x3
sparse_coo = coo_matrix((data, (rows, cols)), shape=(3, 3))

print("Исходная матрица:")
print(sparse_coo.toarray())

print("\nРазрежённая матрица в формате COO:")
print(sparse_coo)

# Output:
# Исходная матрица:
# [[ 0 10 40]
#  [ 0  0 20]
#  [30  0  0]]
 
# Разрежённая матрица в формате COO:
#   (0, 1)	10
#   (1, 2)	20
#   (2, 0)	30
#   (0, 2)	40
```

```python
import numpy as np
import pandas as pd
from scipy.sparse import coo_matrix

# Шаг 1: Создаём базовый NumPy array
data = np.array([[0, 0, 3],
                 [4, 0, 0],
                 [0, 5, 6]])
print("Исходный NumPy array:")
print(data)

# Шаг 2: Преобразуем NumPy array в разрежённую матрицу в формате COO
sparse_matrix = coo_matrix(data)
print("\nРазрежённая матрица (COO формат):")
print(sparse_matrix)

# Шаг 3: Преобразуем разрежённую матрицу обратно в плотный NumPy array
dense_array = sparse_matrix.toarray()
print("\nПлотный NumPy array, полученный из разрежённой матрицы:")
print(dense_array)

# Шаг 4: Создаём DataFrame Pandas на основе плотного массива
df = pd.DataFrame(dense_array, columns=['Column1', 'Column2', 'Column3'])
print("\nDataFrame (Pandas):")
print(df)

# Output:
# Исходный NumPy array:
# [[0 0 3]
#  [4 0 0]
#  [0 5 6]]

# Разрежённая матрица (COO формат):
#   (0, 2)	3
#   (1, 0)	4
#   (2, 1)	5
#   (2, 2)	6

# Плотный NumPy array, полученный из разрежённой матрицы:
# [[0 0 3]
#  [4 0 0]
#  [0 5 6]]

# DataFrame (Pandas):
#    Column1  Column2  Column3
# 0        0        0        3
# 1        4        0        0
# 2        0        5        6
```

```python
import numpy as np

# Задаём два вектора
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Поэлементное умножение: возвращает массив тех же размеров
elementwise = a * b 
print("Поэлементное умножение a * b:")
print(elementwise)

# Задаём две матрицы
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Поэлементное умножение: перемножаются соответствующие элементы
elementwise = A * B  
print("Поэлементное умножение A * B:")
print(elementwise)

# Output:
# Поэлементное умножение a * b:
# [ 4 10 18]
# Поэлементное умножение A * B:
# [[ 5 12]
#  [21 32]]
```

```python
import numpy as np

# Задаём два вектора
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Матричное умножение: возвращает скаляр
dot_product = a @ b  # Результат: 1*4 + 2*5 + 3*6 = 32
print("Скалярное произведение a @ b:")
print(dot_product)

# Задаём две матрицы
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Матричное умножение (стандартное): используется оператор @
matrix_product = A @ B  
print("Матричное умножение A @ B:")
print(matrix_product)

# Output:
# Скалярное произведение a @ b:
# 32
# Матричное умножение A @ B:
# [[19 22]
#  [43 50]]
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Путь к изображению
image_path = "image.jpg"

# Считываем изображение и переводим его в оттенки серого
image = Image.open(image_path).convert("L")
image_np = np.array(image, dtype=float)

# Фильтр понижения яркости: отнимаем константу от всех элементов матрицы
brightness_offset = -50
result_brightness = np.clip(image_np + brightness_offset, 0, 255)

# Фильтр инверсии: вычисляем отрицательное изображение
result_inversion = 255 - image_np

# Пороговая фильтрация: если значение пикселя больше порога, ставим 255, иначе 0
threshold = 128
result_threshold = np.where(image_np > threshold, 255, 0)

# Создаём фигуру с заданными параметрами
fig, axes = plt.subplots(2, 2, figsize=(15, 8), dpi=200)

axes[0, 0].imshow(image_np, cmap='gray')
axes[0, 0].set_title("Исходное изображение")
axes[0, 0].axis('off')

axes[0, 1].imshow(result_brightness, cmap='gray')
axes[0, 1].set_title("Понижение яркости")
axes[0, 1].axis('off')

axes[1, 0].imshow(result_inversion, cmap='gray')
axes[1, 0].set_title("Инверсия изображения")
axes[1, 0].axis('off')

axes[1, 1].imshow(result_threshold, cmap='gray')
axes[1, 1].set_title("Пороговая фильтрация")
axes[1, 1].axis('off')

plt.subplots_adjust(wspace=-0.35, hspace=0.1)
plt.show()
```

```python
# Читаем размерность матрицы без текстовых подсказок
n = int(input())

# Считываем саму матрицу
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

# Изначально предполагаем, что матрица обладает обоими свойствами
is_upper = True
is_lower = True

# Проходим по всем элементам матрицы
for i in range(n):
    for j in range(n):
        if matrix[i][j] != 0:
            if i > j:
                # Если нашли ненулевой элемент ниже главной диагонали, 
                # то матрица не верхнетреугольная
                is_upper = False
            elif i < j:
                # Если нашли ненулевой элемент выше главной диагонали, 
                # то матрица не нижнетреугольная
                is_lower = False

# Выводим результат в зависимости от оставшихся флагов
if is_upper and is_lower:
    print("DIAGONAL")
elif is_upper:
    print("UPPER_TRIANGULAR")
elif is_lower:
    print("LOWER_TRIANGULAR")
else:
    print("OTHER")
```

```python
# 1. Считываем размеры и элементы матрицы A
m, n = map(int, input().split())
A = []
for _ in range(m):
    row = list(map(int, input().split()))
    A.append(row)

# 2. Считываем размеры и элементы матрицы B
h, k = map(int, input().split())
B = []
for _ in range(h):
    row = list(map(int, input().split()))
    B.append(row)

# 3. Проверяем согласованность матриц
if n != h:
    print("NOT_DEFINED")
else:
    # 4. Создаем пустую матрицу C размером m x k, заполненную нулями
    C = [[0] * k for _ in range(m)]

    # 5. Выполняем перемножение
    for i in range(m):              # Проходим по строкам матрицы A
        for j in range(k):          # Проходим по столбцам матрицы B
            for x in range(n):      # Вычисляем скалярное произведение
                C[i][j] += A[i][x] * B[x][j]

    # 6. Выводим результат
    for row in C:
        print(" ".join(map(str, row)))
```

```python
# Функция для перемножения двух квадратных матриц размера n x n
def multiply(A, B, n):
    C = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for x in range(n):
                C[i][j] += A[i][x] * B[x][j]
    return C

# Функция для проверки, состоит ли матрица целиком из нулей
def is_zero(matrix, n):
    for i in range(n):
        for j in range(n):
            if matrix[i][j] != 0:
                return False
    return True

# 1. Читаем данные
n = int(input())
A = []
for _ in range(n):
    row = list(map(int, input().split()))
    A.append(row)

# Если матрица изначально нулевая, степень k = 1
if is_zero(A, n):
    print(1)
else:
    # 2. Начинаем возводить матрицу в степени
    current_matrix = A
    k = 1
    
    # Умножаем до тех пор, пока матрица не станет нулевой
    # Ограничение k <= 100 добавлено из условия задачи, чтобы избежать бесконечного цикла
    while not is_zero(current_matrix, n) and k <= 100:
        current_matrix = multiply(current_matrix, A, n)
        k += 1
        
    # 3. Выводим найденную степень
    print(k)
```

```python
import math

# 1. Читаем размер матрицы
m, n = map(int, input().split())

# 2. Считываем элементы матрицы A
A = []
for _ in range(m):
    row = list(map(int, input().split()))
    A.append(row)

# 3. Нормализуем по столбцам
for j in range(n):
    # Шаг 3.1: Находим сумму столбца для вычисления среднего
    col_sum = 0
    for i in range(m):
        col_sum += A[i][j]
    mean = col_sum / m
    
    # Шаг 3.2: Вычисляем дисперсию и стандартное отклонение
    variance_sum = 0
    for i in range(m):
        variance_sum += (A[i][j] - mean) ** 2
    std_dev = math.sqrt(variance_sum / m)
    
    # Шаг 3.3: Применяем формулу нормализации к каждому элементу столбца
    for i in range(m):
        # Защита от деления на ноль, если все элементы столбца одинаковы
        if std_dev == 0:
            A[i][j] = 0
        else:
            # Вычисляем нормализованное значение
            normalized_value = (A[i][j] - mean) / std_dev
            # int() просто отбрасывает дробную часть (например, 1.9 -> 1; -1.2 -> -1)
            A[i][j] = int(normalized_value)

# 4. Выводим нормализованную матрицу
for row in A:
    print(" ".join(map(str, row)))
```
